```python
# 广搜：
# 从四个边向里面搜索，只要是遇到的’O’，都是可以从外边界走通的，都会用'a’标记 
# 广搜一遍之后，一个一个元素遍历， 如果是’.’说明这个’O’是可以从外界走通的，保留 
# 如果是’O’,说明这是闭塞的，从外界走不通的,替换为’X’。

class Solution:
    def solve(self, board):
        """
        :type board: List[List[str]]
        :rtype: void Do not return anything, modify board in-place instead.
        """
        if len(board) == 0:
            return
        m,n = len(board), len(board[0])
        q = []
        
        def replace(i,j):
            if i < 0 or i > m-1 or j < 0 or j > n-1 or board[i][j] != 'O':
                return
            board[i][j] = 'a'
            q.append((i,j))
        
        def bfs(i,j):
            if board[i][j] == 'O':
                replace(i,j)
            while q:
                cur = q.pop()
                x, y = cur[0], cur[1]
                replace(x-1, y)
                replace(x+1, y)
                replace(x, y-1)
                replace(x, y+1)
                
        for i in range(n):
            bfs(0, i)
            bfs(m-1, i)
        for j in range(1, m-1):
            bfs(j, 0)
            bfs(j, n-1)

        for i in range(m):
            for j in range(n):
                if board[i][j] == 'a':
                    board[i][j] = 'O'
                elif board[i][j] == 'O':
                    board[i][j] = 'X'
        
# board = [['X', 'X', 'X', 'X'],['X', 'O', 'O', 'X'],['X', 'X', 'O', 'X'],['X', 'O', 'X', 'X']]
# board = [["X","O","X","O","X","O"],["O","X","O","X","O","X"],["X","O","X","O","X","O"],["O","X","O","X","O","X"]]
board = [["O","O","O","O","X","X"],["O","O","O","O","O","O"],["O","X","O","X","O","O"],["O","X","O","O","X","O"],
         ["O","X","O","X","O","O"],["O","X","O","O","O","O"]]
for i in range(len(board)):
    print(board[i], '\n')
print('\n')
Solution().solve(board)
for i in range(len(board)):
    print(board[i], "\n")
```
